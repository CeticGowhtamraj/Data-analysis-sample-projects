# Converting the type of date from object to datetime using pd.to_datetime
creating the required columns Month, Year,day of the week


```python
df["Date"]=pd.to_datetime(df["Date"],errors='coerce',dayfirst=True)
df["Month"]=df["Date"].dt.month
df["Year"]=df["Date"].dt.year
df["weekday"]=df["Date"].dt.day_name()
```

#  Sales Trends Over Time

```python
Sales_trends_over_time=df.groupby("Month")["Sales"].sum()
print("Sales trends over time:")
Sales_trends_over_time
```

```python

Sales_trends_over_time.plot(kind="line",marker='o',figsize=(10, 6))
plt.title("Sales Trends Over Time")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

#  Top-Selling Product Lines

```python
Top_Selling_Product= df.groupby("Product line")["Sales"].sum()
Top_Selling_Product
```

```python
Top_Selling_Product.plot(kind="line",figsize=(10, 6),marker="o")
plt.title("Top-Selling Product Linese")
plt.xlabel("products")
plt.ylabel("Sales")
plt.tight_layout()
plt.grid(True)
plt.show()


```

# Profitability Insights

```python
Product_line= df.groupby("Product line")["profit"].sum()
Branch= df.groupby("Branch")["profit"].sum()
Time= df.groupby("Month")["profit"].sum()
Customer_type= df.groupby("Customer type")["profit"].sum()
values=(Product_line,Branch,Time,Customer_type)
values
```

```python
for label, group in {
    "Product line": Product_line,
    "Branch": Branch,
    "Month": Time,
    "Customer type": Customer_type
}.items():
    plt.figure(figsize=(6, 6))
    explode = tuple([0.05] * len(group))
    colors = ["#66b3ff", "#99ff99", "#ffcc99", "#ff6666", "#c2c2f0", "#ffb3e6"][:len(group)]
    
    plt.pie(
        group,
        labels=group.index,
        autopct='%1.1f%%',
        startangle=90,
        explode=explode,
        shadow=True,
        colors=colors,
        wedgeprops={'edgecolor': 'black'}
    )
    plt.title(f"Profit Distribution by {label}")
    plt.tight_layout()
    plt.show()
```

# Lagest Sectors

```python
Product_line_lag= df.groupby("Product line")["Sales"].sum().nsmallest(3)
Branch_lag= df.groupby("Branch")["profit"].sum().nsmallest(3)
Time_lag= df.groupby("Month")["profit"].sum().nsmallest(3)
Customer_type_lag= df.groupby("Customer type")["profit"].sum().nsmallest(3)
print(Product_line_lag,Branch_lag,Time_lag,Customer_type_lag)
```
